# My first line chart
Let's plot a simple line graph of square numbers.

```python
import matplotlib.pyplot as plt
#from now on, I can use plt._() as plotting function instead of writing out the matplotlib line every time.

squares = [1,4,9,16,25]
plt.plot(squares)
plt.show()
```

Here I've entered the data as an array, and plotted a simple line graph. I can improve on the presentation:

```python
plt.plot(squares, linewidth=5)

#Set chart title and label axes
plt.title("Square numbers", fontsize=24, color='white')
plt.xlabel("Value", fontsize=14, color='white')
plt.ylabel("Square of value", fontsize=14, color='white')

#Set size of tick labels.
plt.tick_params(axis='both', labelsize=14, color='white', labelcolor='white')
```

I only gave it the array of y-values to plot, and the plot assumes x values start as 0. Add in input value array to present the correct plot:

```python
x_input = [1,2,3,4,5]

plt.plot(x_input, squares, linewidth=5)

#Set chart title and label axes
plt.title("Square numbers", fontsize=24, color='white')
plt.xlabel("Value", fontsize=14, color='white')
plt.ylabel("Square of value", fontsize=14, color='white')

#Set size of tick labels.
plt.tick_params(axis='both', labelsize=14, color='white', labelcolor='white')
```
